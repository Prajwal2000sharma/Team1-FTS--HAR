```python
#separating the dependent and independent variables
X_train=train_data.drop(columns=['Work'])
Y_train= train_data['Work']
X_test=test_data.drop(columns=['Work'])
Y_test=test_data['Work']
X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
X_test=X_test.interpolate(method='linear',limit_direction='forward')
X_test.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder 
le=LabelEncoder()
Y_train=le.fit_transform(Y_train)
Y_test=le.fit_transform(Y_test)
Y_test
```

```python
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier 
scores = []
for i in range(1, 50):
    knn = KNeighborsClassifier(n_neighbors = i, n_jobs = -1)
    knn.fit(X_train, Y_train)
    Y_pred = knn.predict(X_test)
    scores.append(accuracy_score(Y_test,Y_pred))
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy Score')
xticks = range(1,50)
plt.plot(xticks, scores, color='red', linestyle='solid', marker='o',
         markerfacecolor='blue', markersize=5)
```

as we can see the maximum accuracy comes when number of neighbours is between 15 and 20....we will train on n_neigbours=17

```python
from sklearn.neighbors import KNeighborsClassifier 
knc=KNeighborsClassifier(n_neighbors=17)
knc.fit(X_train,Y_train)
Y_pred=knc.predict(X_test)
print(accuracy_score(Y_test,Y_pred)*100, ' %')
```

```python


from sklearn.metrics import f1_score
f1_score(Y_test,Y_pred,average='weighted')
```
